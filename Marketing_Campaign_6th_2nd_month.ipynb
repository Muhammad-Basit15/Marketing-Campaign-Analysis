**convert columns from object to int**

```python
cols_to_convert = [
    'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
    'MntSweetProducts', 'MntGoldProds', 'Kidhome', 'Teenhome',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'AcceptedCmp1', 'AcceptedCmp2',
    'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'
]

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

**Creating new column**

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
#Total_Spend
df['Total_Spend'] = df['MntWines']+ df['MntFruits']+ df['MntMeatProducts']+df['MntFishProducts']+ df['MntSweetProducts']+ df['MntGoldProds']
```

```python
# Total_Children
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# Total_Purchases
df['Total_Purchases'] = df['NumDealsPurchases']+ df['NumWebPurchases']+ df['NumCatalogPurchases']+ df['NumStorePurchases']
```

```python
# Campaigns_Accepted
df['Campaigns_Accepted'] = df['AcceptedCmp3']+ df['AcceptedCmp4']+ df['AcceptedCmp5']+df['AcceptedCmp1']+ df['AcceptedCmp2']
```

```python
df['Campaigns_Accepted'] = pd.to_numeric(df['Campaigns_Accepted'], errors='coerce')
df['Campaign_ResponseRate'] = df['Campaigns_Accepted'] / 5.0
```

```python
# Is_Responsive (binary flag if accepted any campaign)
df['Is_Responsive'] = df['Campaigns_Accepted'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Age
df['Year_Birth'] = pd.to_datetime(df['Year_Birth'], format='%Y', errors='coerce')
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth'].dt.year
```

```python
# Customer_Lifetime (in days)
df['Customer_Lifetime'] = (datetime.now() - df['Dt_Customer']).dt.days
```

```python
#  Avg_Spend_per_Purchase
df['Total_Spend'] = pd.to_numeric(df['Total_Spend'], errors='coerce')
df['Total_Purchases'] = pd.to_numeric(df['Total_Purchases'], errors='coerce')
df['Avg_Spend_per_Purchase'] = df['Total_Spend'] / df['Total_Purchases']
df['Avg_Spend_per_Purchase'] = df['Avg_Spend_per_Purchase'].replace([float('inf'), float('nan')], 0)
```

```python
df['Camp_cost']= df['z_cost_contact']*5
```

```python
def age_group(age):
    if 21 <= age <= 30:
        return "21-30"
    elif 31 <= age <= 40:
        return "31-40"
    elif 41 <= age <= 50:
        return "41-50"
    elif 51 <= age <= 60:
        return "51-60"
    elif 61 <= age <= 70:
        return "61-70"
    elif 71 <= age <= 80:
        return "71-80"
    elif 81 <= age <= 90:
        return "81-90"
    elif 91 <= age <= 100:
        return "91-100"
    else:
        return "Other"

df['Age_Group'] = df['Age'].apply(age_group)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
df['Possible_revenue'] = np.where(
    df['Is_Responsive'] == 1,
    (df['Campaigns_Accepted'] * df['Z_Revenue']) + (df['Campaigns_Accepted'] * df['Avg_Spend_per_Purchase']),
    0
)
```

```python
df['Response_status'] = df['Is_Responsive'].apply(lambda x: 'Accepted' if x == 1 else 'Not Accepted')
```

```python
def get_salary_range(income):
    if income < 1000:
        return "Below 1k"
    elif income <= 20000:
        return "1k-20k"
    elif income <= 40000:
        return "20k-40k"
    elif income <= 60000:
        return "40k-60k"
    elif income <= 80000:
        return "60k-80k"
    elif income <= 100000:
        return "80k-100k"
    elif income <= 120000:
        return "100k-120k"
    elif income <= 140000:
        return "120k-140k"
    elif income <= 150000:
        return "140k-150k"
    else:
        return "Above 150k"


df['Salary_Range'] = df['Income'].apply(get_salary_range)
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
```
